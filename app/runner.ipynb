## Runner
### To try differnt paramter of stock_analyser

### Set all parameter in below block

`MOVING_AVERAGE`
-  if set, will calculate extrema base on the moving average
- options: 'ma', 'ema', 'dma'

`MOVING_AVERAGE_PERIOD`
- set day period of moving average
- e.g. `MOVING_AVERAGE`='ma'; `MOVING_AVERAGE_PERIOD`=9 -> i.e. calculate extrema base on MA9

`SMOOTH`: smoothen the curve before calculating extrema or not

`SMOOTH_EXTEND`: extend of smoothening (only apply if SMOOTH==True)

`WINDOW`: window on locating extrema on actual price base on ma

`SHOW_ALL_VERTEX`: if set true, will override other setting, and just locate all vertex base on actual stock price (no smoothening and approximating)

```python
import stock_analyser as sa
```

```python
# Set Parameter

STOCK = 'NVDA'
START = '2023-04-20'
END = '2023-07-20'
MOVING_AVERAGE = 'ema'
MOVING_AVERAGE_PERIOD = 5
SMOOTH = False          # smoothen the curve before calculating extrema
SMOOTH_EXTEND = 0       # extend of smoothening (only apply if SMOOTH==True)
WINDOW = 5              # window on locating extrema on actual price base on ma
SHOW_ALL_VERTEX = False # if set true, will override other setting, and just locate all vertex base on actual stock price
```
